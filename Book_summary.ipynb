```python
%%sql
CREATE TABLE users (
    id            SERIAL          PRIMARY KEY,
    full_name     VARCHAR(150)    NOT NULL,
    email         VARCHAR(255)    UNIQUE NOT NULL,
    password_hash TEXT            NOT NULL,
    created_at    TIMESTAMP       DEFAULT CURRENT_TIMESTAMP,
    last_login    TIMESTAMP,
    is_active     BOOLEAN         DEFAULT TRUE
);
```

```python
%%sql
select * from users
```

```python
%%sql
CREATE TABLE user_preferences (
    id              SERIAL          PRIMARY KEY,
    user_id         INTEGER         NOT NULL,
    goal            VARCHAR(100),
    background      VARCHAR(100),
    style           VARCHAR(50),
    tone            VARCHAR(50),
    genre_1         VARCHAR(100),
    genre_2         VARCHAR(100),
    profile_summary TEXT,
    created_at      TIMESTAMP       DEFAULT CURRENT_TIMESTAMP,
    updated_at      TIMESTAMP       DEFAULT CURRENT_TIMESTAMP,
 
    CONSTRAINT fk_preferences_user
        FOREIGN KEY (user_id)
        REFERENCES users(id)
        ON DELETE CASCADE
);
```

```python
%%sql
CREATE TABLE books (
    id             SERIAL          PRIMARY KEY,
    title          VARCHAR(500)    NOT NULL,
    author         VARCHAR(300),
    genre          VARCHAR(200),
    cover_image    TEXT,
    description    TEXT,
    published_year VARCHAR(10),
    created_at     TIMESTAMP       DEFAULT CURRENT_TIMESTAMP,
 
    CONSTRAINT uq_books_title_author
        UNIQUE (title, author)
);
```

```python
%%sql
CREATE TABLE summaries (
    id SERIAL PRIMARY KEY,

    book_id INTEGER NOT NULL REFERENCES books(id) ON DELETE CASCADE,

    language VARCHAR(10) NOT NULL,

    whats_inside TEXT,
    youll_learn JSONB,
    key_points JSONB,
    about_author TEXT,
    quote TEXT,
    conclusion TEXT,
    cache_key TEXT NOT NULL,

    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,

    UNIQUE (book_id, language, cache_key)
);
```

```python
%%sql
DROP TABLE summaries cascade
```

```python
%%sql
CREATE TABLE chat_history (
    id SERIAL PRIMARY KEY,

    user_id INTEGER REFERENCES users(id) ON DELETE CASCADE,
    book_id INTEGER REFERENCES books(id) ON DELETE CASCADE,
    summary_id INTEGER REFERENCES summaries(id) ON DELETE SET NULL,

    user_message TEXT,
    ai_response TEXT,

    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
```

```python
%%sql
drop table chat_history cascade
```

```python
%%sql
CREATE TABLE key_point_details (
    id           SERIAL    PRIMARY KEY,
    book_id      INTEGER   NOT NULL,
    reading_time INTEGER   NOT NULL,
    point_index  INTEGER   NOT NULL,
    title        TEXT      NOT NULL,
    full_detail  JSONB     NOT NULL,
    created_at   TIMESTAMP DEFAULT CURRENT_TIMESTAMP,

    CONSTRAINT uq_book_point
        UNIQUE (book_id, reading_time, point_index),

    CONSTRAINT fk_keypoint_book
        FOREIGN KEY (book_id)
        REFERENCES books(id)
        ON DELETE CASCADE
);
```

```python
%%sql
CREATE INDEX idx_summaries_book_lang 
ON summaries(book_id, language);

CREATE INDEX idx_summaries_cache 
ON summaries(cache_key);

CREATE INDEX idx_chat_user 
ON chat_history(user_id);
```
